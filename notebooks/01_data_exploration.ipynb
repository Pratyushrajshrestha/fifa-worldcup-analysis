```python
#from above result we found that there is a runners-up column ,which can be problem due to -
#changing it 
df_cups = df_cups.rename(columns={'Runners-Up': 'RunnersUp'})
print(df_cups.columns)
df_cups.dtypes
```

```python
df_cups['Attendance'].dtype
df_cups.Attendance.isnull().sum()
df_cups['Attendance'].head(10)
```

```python
# since the attendance data are in invalid way 1.865.753 its an error
# so removing the dots
df_cups['Attendance'] = df_cups['Attendance'].str.replace('.',' ',regex=False)

# attendance is in str so changing it to number
df_cups['Attendance'] = pd.to_numeric(df_cups['Attendance'], errors='coerce')
df_cups['Attendance'].head(10)
```

```python
print(df_cups['Attendance'].dtype)
df_cups.isnull().sum()
df_cups['Attendance'].head(10)
```

```python
df_cups = pd.read_csv('../data/raw/historical/WorldCups.csv', encoding='unicode_escape')
df_cups['Attendance'].head(10)
df_cups['Attendance'] = df_cups['Attendance'].str.strip()
df_cups['Attendance'] = df_cups['Attendance'].str.replace('.', '', regex=False) #removed the .
df_cups['Attendance'].head(10)
df_cups['Attendance'] = pd.to_numeric(df_cups['Attendance'], errors='coerce') #str to int/float
print(df_cups['Attendance'].dtype)
df_cups.isnull().sum()
```

```python
# created a new csv file in processed file where the clean data of worlcups.csv are kept    
df_cups.to_csv('../data/processed/cups_clean.csv', index=False)
```

```python
print(df_matches.shape)
df_matches.isnull().sum()
```

```python
# droping empty rows
df_matches = df_matches.dropna(how='all')
df_matches.shape
```

```python
df_matches.isnull().sum()
```

```python
#since two missing data of attendance we remove it
df_matches = df_matches.dropna(subset='Attendance')
df_matches.isnull().sum()
```

```python
df_matches.dtypes
# since year type is float which is not good so change it to int

df_matches['Year'] = df_matches['Year'].astype(int)
```

```python
df_matches.columns.tolist()
```

```python
# since spaces can make problems we use _ instead of space
df_matches.columns = df_matches.columns.str.strip().str.lower().str.replace(' ','_').str.replace('-','_')
print(df_matches.columns.tolist())
```

```python
#now we are creating a new csv file in processed file where the clean data of worldcupmatches.csv are kept
df_matches.to_csv('../data/processed/matches_clean.csv', index=False)
```
